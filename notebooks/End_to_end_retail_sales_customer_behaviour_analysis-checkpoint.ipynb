# End-to-End Retail Sales & Customer Behaviour Analysis with Seasonal and External Factors


### Project Description

Retail businesses rely on data-driven insights to understand customer purchasing behaviour, identify seasonal trends, and optimize sales performance. This project presents an end-to-end data analytics workflow focused on analyzing retail transaction data to understand sales performance, order volume, customer purchasing behaviour, and seasonal trends. The primary dataset consists of transactional data from a UK-based online retail store. Using python in Jupyter Notebook, the project covers data cleaning, feature engineering, and exploratory data analysis to uncover patterns in revenue, product perfomance, and customer activity over time. External contextual datasets including public holidays and monthly weather averages, are incoporated to assess the influence of seasonal and environmental factors on sales trends. Key insights are communicated through statistical analysis and visual exploration, while final business facing dashboards are developed in Power BI to support data-driven decision-making. 

### Primary Objectives

- To clean, prepare, and structure retail transaction data for analysis using python
- To analyze sales trends, revenue distribution, and customer purchasing behaviour over time
- To identify seasonal patterns in retail perfomance using time-based features such as month and quarter
- To assess the impact of external factors including public holidays and weather conditions, on sales trends.

### Data Sources & Description

- Online retail transactional data (UK-based online store)
- UK public holidays
- Monthly average weather data(Heathrow weather station)
- Cunsumer Price Index(CPI) data

#### Performing Exploratory Data Analysis on transactional data

```python
sales_df = uk_retail_df
#checking quantity distibution
sales_df['Quantity'].describe()
```

```python
#checking for any negative values in quantities
sales_df[sales_df['Quantity']<0].shape
```

Negative values, which represent product returns or cancelled transactions, were removed to ensure the analysis reflects actual sales activity. 

```python
#removing negative quantities or cancellation/returns
sales_df = sales_df[sales_df['Quantity']>0].copy()
```

```python
#inspecting extreme positive outliers
sales_df['Quantity'].sort_values(ascending=False).head(20)
```

```python
#visualising this in a box and whisker to determine the method to use in dealing with extreme positive outliers.
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 3))
sns.boxplot(x=sales_df['Quantity'])
plt.xscale('log')
plt.title("Boxplot of Quantity per Transaction")
plt.xlabel("Quantity")
plt.show()
```

```python
#quantifying the total number of outliers. 
Q1 = sales_df['Quantity'].quantile(0.25)
Q3 = sales_df['Quantity'].quantile(0.75)
IQR = Q3 - Q1
Q1, Q3, IQR
```

```python
#Defining box plot outlier bounds
lower_bound = Q1 - 1.5*IQR
upper_bound = Q3 + 1.5*IQR
lower_bound, upper_bound
```

```python
#counting outliers outside the box plot
outliers = sales_df[(sales_df['Quantity']<lower_bound) | (sales_df['Quantity']>upper_bound)]
outliers.shape[0]
```

```python
#seeing the proportion of outliers
outlier_percentage = outliers.shape[0]/sales_df.shape[0] *100
outlier_percentage
```

Using the boxplot (IQR) method, observations outside the range Q1-1.5*IQR and Q3 + 1.5*IQR were identifiedas ouliers. These represented approximately 10% of the total transactions, indicating the presence of extreme purchasing behaviour.

```python
#dealing with outliers using percentage based removal(the 99th percentile approach)
#computing the threshold
quantity_99 = sales_df["Quantity"].quantile(0.99)
quantity_99
```

```python
#removing extreme rows
sales_df = sales_df[sales_df['Quantity']<= quantity_99].copy()
sales_df['Quantity'].describe()
```

```python
#replotting the box plot
sns.boxplot(x=sales_df['Quantity'])
plt.show()
```

The IQR-based box plot method identified approximately 9.8% of transactionsas outliers a highly right-skewed distribution. Toavoid removing a substantial portion of valid sales data, a percentile-based approach as adopted , with extreme values above the 99th percentile treated as outliers. 

Although extreme values were removed using a percentile-based approacg, some observations remain outside the boxplot whiskers. This is expected, as the box plot uses the IQR rule, which flags value far from the central distribution. These remaining values represent legitimate bulk purchases rather than data errors.

Unit prices were inspected for invalid and extreme values. Observations with zero or negative prices were remove . Due to hughly  right-skewed distribution, a percentile based approach was used to handle extreme values, ensuring that legitimate high-priced items were retained while limiting the influence of extreme observations

```python
#creating the revenue column 
sales_df['Revenue'] = sales_df['Quantity']*sales_df['Price']
sales_df[['Quantity', 'Price', 'Revenue']].head(5)
```

```python
sales_df['Revenue'].describe()
```

```python
#visualizing revenue distribution
sns.histplot(sales_df['Revenue'], bins = 25)
plt.title('Revenue distribution')
plt.show()
```

```python
sns.boxplot( x= sales_df[sales_df['Revenue'] <= sales_df['Revenue'].quantile(0.95)]['Revenue'])
plt.title('Revenue boxplot (<=95th percentile)')
plt.show()
```

```python
#Handling extreme values
revenue_cap = sales_df['Revenue'].quantile(0.995)
sales_df['Revenue'] = sales_df['Revenue'].clip(upper=revenue_cap)
sales_df['Revenue'].describe()
```

### Cleaning the Monthly Average Weather Data(Heathrow weather station)

```python
monthly_average_weather = pd.read_csv(r"C:\Users\ASUS\Desktop\PL 300 Documents\MET Office Weather Data.csv")
monthly_average_weather.head(3)
```

```python
#filtering data for the year between 2009 and 2011 to fit the date ranges we have in the transactional dataset
monthly_average_weather = monthly_average_weather[((monthly_average_weather["year"] >= 2009) & 
                                                   (monthly_average_weather["year"] <=2011))]
#filtering data for the months from December 2009 to December 2011 to fit the month ranges we have in the transactional dataset
monthly_average_weather = monthly_average_weather[(monthly_average_weather["year"] > 2009) | 
                                                  ((monthly_average_weather["year"] ==2009)& 
                                                   (monthly_average_weather["month"]>=12))]
#filtering the data where station is Heathrow
monthly_average_weather = monthly_average_weather[(monthly_average_weather["station"] == "heathrow")]
#rename column af to air frost for clarity
monthly_average_weather = monthly_average_weather.rename(columns ={"af":"air frost"})
#conerting the year and month column column to integers from float values
monthly_average_weather["year"] = monthly_average_weather["year"].astype("int")
monthly_average_weather["month"] = monthly_average_weather["month"].astype("int")
monthly_average_weather.head(5)
```

```python
#checking duplicates values
monthly_average_weather.duplicated().sum()
```

```python
#checking missing values
monthly_average_weather.isna().sum()
```

```python
#basic stats per column
monthly_average_weather[['tmax','tmin','air frost','rain','sun']].describe()
```

```python
#checking correlations between weather variables
sns.heatmap(monthly_average_weather[['tmax','tmin','air frost', 'rain', 'sun']].corr(), annot=True, cmap= 'coolwarm')
plt.title('Correlation between weather variables')
plt.show()
```

### Cleaning the monthly cpi index data

```python
monthly_cpi_index = pd.read_csv(r"C:\Users\ASUS\Desktop\PL 300 Documents\series-151225.csv")
monthly_cpi_index.head(6)
```

```python
#keeping only the month and year in the year column
mask = monthly_cpi_index["year"].astype(str).str.match(r'^\d{4}\s?[A-Za-z]{3,}$')
monthly_cpi_index=monthly_cpi_index[mask].reset_index(drop=True)
monthly_cpi_index.head(5)
```

```python
#normalizing the date format
monthly_cpi_index['year_month'] = pd.to_datetime(monthly_cpi_index['year'],
format = '%Y %b')
#clean up by dropping the year column
monthly_cpi_index = monthly_cpi_index.drop(columns=['year'])
monthly_cpi_index['year_month'].dt.strftime('%Y %b').unique()
monthly_cpi_index['year_month'] = monthly_cpi_index['year_month'].dt.to_period('M')

monthly_cpi_index.tail(5)
```
